### **An√°lisis de visitas**:

##### Paso 2.1 Determinar los usuarios activos diarios, semanales y mensuales.

```python
# An√°lisis de DAU, WAU y MAU
# 1. Cargar datos de visitas
visits = pd.read_csv('visits_log_us.csv')

# 2. Renombrar columnas correctamente
visits = visits.rename(columns={
    'Start Ts': 'start_ts',
    'End Ts': 'end_ts',
    'Device': 'device',
    'Source Id': 'source_id',
    'Uid': 'uid'
})

# 3. Parsear fechas y optimizar tipos
visits['start_ts'] = pd.to_datetime(visits['start_ts'])
visits['end_ts'] = pd.to_datetime(visits['end_ts'])
visits['device'] = visits['device'].astype('category')
visits['source_id'] = visits['source_id'].astype('category')
visits['uid'] = visits['uid'].astype('category')

# 4. Agregar columna de fecha
visits['date'] = visits['start_ts'].dt.date

# 5. C√°lculo de DAU
dau = visits.groupby('date')['uid'].nunique().reset_index(name='DAU')

# 6. C√°lculo de WAU (usando lunes de cada semana)
visits['week_start'] = visits['start_ts'].dt.to_period('W').apply(lambda r: r.start_time.date())
wau = visits.groupby('week_start')['uid'].nunique().reset_index(name='WAU')

# 7. C√°lculo de MAU (primer d√≠a de cada mes)
visits['month_start'] = visits['start_ts'].dt.to_period('M').apply(lambda r: r.start_time.date())
mau = visits.groupby('month_start')['uid'].nunique().reset_index(name='MAU')

# 8. Graficar m√©tricas
plt.figure(figsize=(12, 4))
sns.lineplot(data=dau, x='date', y='DAU')
plt.title('DAU: Usuarios Activos Diarios')
plt.xlabel('Fecha')
plt.ylabel('DAU')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 4))
sns.lineplot(data=wau, x='week_start', y='WAU')
plt.title('WAU: Usuarios Activos Semanales')
plt.xlabel('Inicio de Semana')
plt.ylabel('WAU')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 4))
sns.lineplot(data=mau, x='month_start', y='MAU')
plt.title('MAU: Usuarios Activos Mensuales')
plt.xlabel('Inicio de Mes')
plt.ylabel('MAU')
plt.tight_layout()
plt.show()

# 9. Mostrar primeros registros de cada serie
print("Primeras filas de DAU:")
print(dau.head(), "\n")
print("Primeras filas de WAU:")
print(wau.head(), "\n")
print("Primeras filas de MAU:")
print(mau.head())
```

He calculado y graficado:

1- DAU (Daily Active Users): Usuarios √∫nicos por d√≠a.

2- WAU (Weekly Active Users): Usuarios √∫nicos por semana, usando el lunes como inicio de semana.

3- MAU (Monthly Active Users): Usuarios √∫nicos por mes, usando el primer d√≠a como referencia.

Los primeros registros muestran:

1- DAU arranca en 605 usuarios el 1 de junio de 2017.

2- WAU inicia en 2 021 usuarios la semana del 29 de mayo de 2017.

3- MAU comienza con 13 259 usuarios en junio de 2017.

#### Paso 2.3 - Calcular la duraci√≥n de cada sesi√≥n.

```python

# Cargar y preparar datos
visits = pd.read_csv('visits_log_us.csv')
visits = visits.rename(columns={'Start Ts': 'start_ts', 'End Ts': 'end_ts'})
visits['start_ts'] = pd.to_datetime(visits['start_ts'])
visits['end_ts'] = pd.to_datetime(visits['end_ts'])
visits['date'] = visits['start_ts'].dt.date

# Calcular duraci√≥n de cada sesi√≥n en segundos
visits['duration_sec'] = (visits['end_ts'] - visits['start_ts']).dt.total_seconds()

# Resumen estad√≠stico de duraci√≥n
duration_summary = visits['duration_sec'].describe()

# Duraci√≥n media diaria de sesi√≥n
daily_avg_duration = visits.groupby('date')['duration_sec'].mean().reset_index(name='avg_duration_sec')

# Gr√°fico 1: Histograma de la duraci√≥n de sesiones
plt.figure(figsize=(10, 4))
sns.histplot(visits['duration_sec'], bins=50, kde=True)
plt.title('Distribuci√≥n de la Duraci√≥n de Sesiones (segundos)')
plt.xlabel('Duraci√≥n (s)')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()

# Gr√°fico 2: Evoluci√≥n de la duraci√≥n media diaria
plt.figure(figsize=(10, 4))
sns.lineplot(data=daily_avg_duration, x='date', y='avg_duration_sec')
plt.title('Duraci√≥n Media Diaria de Sesi√≥n')
plt.xlabel('Fecha')
plt.ylabel('Duraci√≥n Media (s)')
plt.tight_layout()
plt.show()

# Mostrar estad√≠sticas y primeras filas del promedio diario
print("Resumen estad√≠stico de la duraci√≥n de sesiones (segundos):")
print(duration_summary)
print("\nPrimeras filas de la duraci√≥n media diaria:")
print(daily_avg_duration.head())
```

#### Paso 2.4 - Analizar la frecuencia de retorno de los usuarios.

```python

# Cargar y preparar datos
visits = pd.read_csv('visits_log_us.csv')
visits = visits.rename(columns={'Start Ts': 'start_ts', 'Uid': 'uid'})
visits['start_ts'] = pd.to_datetime(visits['start_ts'])

# Ordenar y calcular d√≠as de retorno
visits = visits.sort_values(['uid', 'start_ts'])
visits['prev_time'] = visits.groupby('uid')['start_ts'].shift(1)
visits['return_diff_days'] = (visits['start_ts'] - visits['prev_time']).dt.days

# Filtrar diffs v√°lidos
diffs = visits['return_diff_days'].dropna()

# Histograma de retornos hasta 30 d√≠as
plt.figure(figsize=(10, 4))
sns.histplot(diffs[diffs <= 30], bins=30, kde=False)
plt.title('Distribuci√≥n de D√≠as entre Visitas (‚â§30 d√≠as)')
plt.xlabel('D√≠as hasta la siguiente visita')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()

# C√°lculo de retenci√≥n en ventanas 1, 7 y 30 d√≠as
first_visits = visits.groupby('uid')['start_ts'].min().reset_index(name='first_visit')
min_return = visits.groupby('uid')['return_diff_days'].min().reset_index(name='min_diff')
ret = first_visits.merge(min_return, on='uid')
total_users = ret.shape[0]

windows = [1, 7, 30]
ret_data = []
for w in windows:
    count = (ret['min_diff'] <= w).sum()
    ret_data.append({
        'window_days': w,
        'users_returned': int(count),
        'pct_returned': round(count / total_users * 100, 2)
    })

ret_df = pd.DataFrame(ret_data)
```

En la distribuci√≥n de d√≠as entre visitas (‚â§30 d√≠as) se observa que la gran mayor√≠a de retornos ocurre en los primeros d√≠as tras una sesi√≥n, con un pico en el primer d√≠a:

Aproximadamente 9.9 % de los usuarios regresan al d√≠a siguiente (1 d√≠a).

El 13.1 % lo hace dentro de una semana (‚â§7 d√≠as).

Y el 16.9 % regresa dentro de un mes (‚â§30 d√≠as).

Estos resultados indican que, aunque un porcentaje peque√±o vuelve r√°pidamente, la curva de retorno es muy pronunciada al inicio y luego se extiende con cola larga.

### Paso 2.5 - Determine cu√°ndo los usuarios comienzan a comprar, centr√°ndose en el tiempo transcurrido entre el registro y la conversi√≥n (p. ej., Conversi√≥n 0d, Conversi√≥n 1d).

```python


# Cargar datos de costos
costs = pd.read_csv('costs_us.csv')

# Renombrar y parsear fechas
costs = costs.rename(columns={'dt': 'date', 'costs': 'cost', 'source_id': 'source_id'})
costs['date'] = pd.to_datetime(costs['date'])
costs['source_id'] = costs['source_id'].astype('category')
costs['cost'] = costs['cost'].astype('float32')

# 1. Gasto total de marketing
total_spend = costs['cost'].sum()

# 2. Gasto por fuente de adquisici√≥n
spend_by_source = costs.groupby('source_id')['cost'].sum().reset_index()
spend_by_source = spend_by_source.sort_values(by='cost', ascending=False)

# 3. Gasto a lo largo del tiempo (mensual)
monthly_spend = (
    costs
    .set_index('date')
    .resample('M')['cost']
    .sum()
    .reset_index(name='monthly_cost')
)
monthly_spend['month'] = monthly_spend['date'].dt.to_period('M').dt.to_timestamp()

# Gr√°ficos
plt.figure(figsize=(10, 5))
sns.barplot(data=spend_by_source, x='source_id', y='cost')
plt.title('Gasto por Fuente de Adquisici√≥n')
plt.xlabel('Fuente de Adquisici√≥n')
plt.ylabel('Gasto (USD)')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 5))
sns.lineplot(data=monthly_spend, x='month', y='monthly_cost')
plt.title('Gasto Mensual de Marketing')
plt.xlabel('Mes')
plt.ylabel('Gasto (USD)')
plt.tight_layout()
plt.show()
```

### Paso 2.7 - Determine el tama√±o promedio de compra.

### Paso 2.8 - Calcule el Valor del Tiempo de Vida del Cliente (LTV).

```python

# 1. Cargar datos
visits = pd.read_csv('visits_log_us.csv')
orders = pd.read_csv('orders_log_us.csv')

# 2. Renombrar columnas y parsear fechas
visits = visits.rename(columns={'Start Ts': 'start_ts', 'Source Id': 'source_id', 'Uid': 'uid'})
visits['start_ts'] = pd.to_datetime(visits['start_ts'])
orders = orders.rename(columns={'Buy Ts': 'buy_ts', 'Revenue': 'revenue', 'Uid': 'uid'})
orders['buy_ts'] = pd.to_datetime(orders['buy_ts'])

# 3. Obtener fuente de adquisici√≥n y fecha de primer visita por usuario
first_visits = visits.sort_values('start_ts').groupby('uid').first().reset_index()[['uid', 'source_id', 'start_ts']]

# 4. Calcular revenue total por usuario
revenue_per_user = orders.groupby('uid')['revenue'].sum().reset_index(name='total_revenue')

# 5. Unir adquisici√≥n con revenue (usuarios sin pedidos tendr√°n NaN -> 0)
user_ltv = first_visits.merge(revenue_per_user, on='uid', how='left')
user_ltv['total_revenue'] = user_ltv['total_revenue'].fillna(0)

# 6. Calcular LTV medio global y por fuente
overall_ltv = user_ltv['total_revenue'].mean()
ltv_by_source = user_ltv.groupby('source_id')['total_revenue'].mean().reset_index(name='avg_ltv')

# 7. Mostrar LTV global
print(f"LTV medio global (2017-2018): ${overall_ltv:.2f} USD")

# 8. Graficar LTV por fuente
plt.figure()
plt.bar(ltv_by_source['source_id'].astype(str), ltv_by_source['avg_ltv'])
plt.title('LTV Medio por Fuente de Adquisici√≥n')
plt.xlabel('Fuente de Adquisici√≥n')
plt.ylabel('LTV Medio (USD)')
plt.tight_layout()
plt.show()
```

### Paso 2.10 - Determine el coste de adquisici√≥n de clientes (CAC) para cada fuente publicitaria.

```python

# 1. Cargar datos
visits = pd.read_csv('visits_log_us.csv')
visits = visits.rename(columns={'Start Ts': 'start_ts', 'Source Id': 'source_id', 'Uid': 'uid'})
visits['start_ts'] = pd.to_datetime(visits['start_ts'])

costs = pd.read_csv('costs_us.csv')
costs = costs.rename(columns={'dt': 'date', 'costs': 'cost', 'source_id': 'source_id'})
costs['date'] = pd.to_datetime(costs['date'])
costs['cost'] = costs['cost'].astype('float32')

# 2. Filtrar visitas y determinar primera fuente de adquisici√≥n por usuario
mask = (visits['start_ts'] >= '2017-01-01') & (visits['start_ts'] <= '2018-12-31')
visits_period = visits.loc[mask]
first_visits = (
    visits_period
    .sort_values('start_ts')
    .groupby('uid', as_index=False)
    .first()[['uid', 'source_id']]
)

# 3. Contar usuarios adquiridos por fuente
users_per_source = (
    first_visits
    .groupby('source_id')
    .size()
    .reset_index(name='num_users')
)

# 4. Sumar gasto por fuente
spend_by_source = (
    costs
    .groupby('source_id')['cost']
    .sum()
    .reset_index(name='total_cost')
)

# 5. Calcular CAC = total_cost / num_users
cac_df = spend_by_source.merge(users_per_source, on='source_id')
cac_df['cac'] = cac_df['total_cost'] / cac_df['num_users']

# 6. Gr√°fico de CAC
plt.figure(figsize=(8, 4))
plt.bar(cac_df['source_id'].astype(str), cac_df['cac'])
plt.title('Coste de Adquisici√≥n de Clientes (CAC) por Fuente')
plt.xlabel('Fuente de Adquisici√≥n')
plt.ylabel('CAC (USD)')
plt.tight_layout()
plt.show()
```

Observaciones:

Las fuentes 4, 9 y 10 tienen el CAC m√°s bajo (~$0.85 USD), lo cual indica eficiencia en la adquisici√≥n de usuarios.

Las fuentes 1, 2 y 3 presentan CAC m√°s alto (> $2 USD).

### Paso 2.11 - Eval√∫e la rentabilidad de las inversiones (ROMI).

```python

# 1. Cargar datos
visits = pd.read_csv('visits_log_us.csv')
visits = visits.rename(columns={'Start Ts': 'start_ts', 'Source Id': 'source_id', 'Uid': 'uid'})
visits['start_ts'] = pd.to_datetime(visits['start_ts'])

orders = pd.read_csv('orders_log_us.csv')
orders = orders.rename(columns={'Buy Ts': 'buy_ts', 'Revenue': 'revenue', 'Uid': 'uid'})
orders['buy_ts'] = pd.to_datetime(orders['buy_ts'])

costs = pd.read_csv('costs_us.csv')
costs = costs.rename(columns={'dt': 'date', 'costs': 'cost', 'source_id': 'source_id'})
costs['date'] = pd.to_datetime(costs['date'])

# 2. Determinar primera fuente por usuario
first_visits = visits.sort_values('start_ts').groupby('uid', as_index=False).first()[['uid', 'source_id']]

# 3. Calcular LTV medio por fuente
revenue_per_user = orders.groupby('uid')['revenue'].sum().reset_index(name='total_revenue')
user_ltv = first_visits.merge(revenue_per_user, on='uid', how='left').fillna(0)
ltv_by_source = user_ltv.groupby('source_id')['total_revenue'].mean().reset_index(name='avg_ltv')

# 4. Calcular CAC por fuente
users_per_source = first_visits.groupby('source_id')['uid'].nunique().reset_index(name='num_users')
costs_by_source = costs.groupby('source_id')['cost'].sum().reset_index(name='total_cost')
cac_df = costs_by_source.merge(users_per_source, on='source_id')
cac_df['cac'] = cac_df['total_cost'] / cac_df['num_users']

# 5. Calcular ROMI: (avg_ltv - cac) / cac
romi_df = ltv_by_source.merge(cac_df, on='source_id')
romi_df['romi'] = (romi_df['avg_ltv'] - romi_df['cac']) / romi_df['cac']

# 6. Visualizar ROMI
plt.figure(figsize=(8, 4))
sns.barplot(data=romi_df, x='source_id', y='romi')
plt.title('ROMI por Fuente de Adquisici√≥n')
plt.xlabel('Fuente de Adquisici√≥n')
plt.ylabel('ROMI')
plt.axhline(0, color='gray', linestyle='--')
plt.tight_layout()
plt.show()
```

    Fuente 1: ROMI ~49% ‚Äî inversi√≥n muy rentable.

    Fuente 2: ROMI ~9% ‚Äî rentable pero justo al l√≠mite.

    Fuente 9: ROMI ~3% ‚Äî ligero retorno positivo.
